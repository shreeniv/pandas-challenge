```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = len(school_data_complete["school_name"].unique())
total_schools
# Calculate the total number of students
total_students = school_data_complete["Student ID"].count()
total_students
# Calculate the total budget
total_budget = school_data["budget"].sum()
total_budget
# Calculate the average math score
avg_math_score = school_data_complete["math_score"].mean()
avg_math_score
# Calculate the average reading score
avg_reading_score = school_data_complete["reading_score"].mean()
avg_reading_score
# Calculate passing math score
pass_math_count = school_data_complete[(school_data_complete["math_score"]>=70)].count()["student_name"]
pass_math_count
# Calculate passing reading score
pass_read_count = school_data_complete[(school_data_complete["reading_score"]>=70)].count()["student_name"]
pass_read_count
# Calculate the percentage of students with a passing math score (70 or greater)
percent_pass_math = (pass_math_count/total_students)*100
percent_pass_math
# Calculate the percentage of students with a passing reading score (70 or greater)
percent_pass_read = (pass_read_count/total_students)*100
percent_pass_read
# Calculate the number passed math **and** reading 
overall_pass_count = school_data_complete[(school_data_complete["math_score"]>=70)&(school_data_complete["reading_score"]>=70)].count()["student_name"]
overall_pass_count
# Calculate the percentage of students who passed math **and** reading (% Overall Passing)
overall_percent = (overall_pass_count/total_students)*100
overall_percent
# Create a dataframe to hold the above results
district_summary = pd.DataFrame({"Total School": [total_schools],
                                "Total Students": [total_students],
                                "Total Budget": [total_budget],
                                 "Average Math Score": [avg_math_score],
                                 "Average Reading Score": [avg_reading_score],
                                 "Percent Pass in Math": [percent_pass_math],
                                 "Percent Pass in Reading": [percent_pass_read],
                                 "Percent Overall Passing": [overall_percent]})
# Optional: give the displayed data cleaner formatting
district_summary["Total Budget"]=district_summary["Total Budget"].map("${:,.2f}".format)
district_summary["Total Students"]=district_summary["Total Students"].map("{:,}".format)
# Display Dataframe
district_summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Fetch School type for School Name
school_type = school_data.set_index("school_name")["type"]
school_type
# Fetch Total Students count for School Name
total_stu_count = school_data_complete["school_name"].value_counts()
total_stu_count
# Fetch Total students budget for School Name
total_school_budget = school_data_complete.groupby(["school_name"]).mean()["budget"]
total_school_budget
# Fetch per student budget for School Name
per_stu_budget = total_school_budget/total_stu_count
per_stu_budget
# Fetch Average Math Score for School Name
avg_math_score = school_data_complete.groupby(["school_name"]).mean()["math_score"]
avg_math_score
# Fetch Average Reading Score for School Name
avg_read_score = school_data_complete.groupby(["school_name"]).mean()["reading_score"]
avg_read_score 
# Fetch Passing Math Score 
pass_math_count = school_data_complete[(school_data_complete["math_score"]>=70)]
pass_math_count
# Fetch Passing Reading Score 
pass_read_count = school_data_complete[(school_data_complete["reading_score"]>=70)]
pass_read_count
# Fetch Percentage Passing Math Score for School Name 
per_school_pass_math = (pass_math_count.groupby("school_name").count()["student_name"]/total_stu_count)*100
per_school_pass_math
# Fetch Percentage Passing Reading Score for School Name
per_school_pass_read = (pass_read_count.groupby("school_name").count()["student_name"]/total_stu_count)*100
per_school_pass_read
# Fetch Overall passing count
overall_pass_count = school_data_complete[(school_data_complete["math_score"] >= 70)
                                          & (school_data_complete["reading_score"] >= 70)]
overall_pass_count
# Fetch Overall Passing percentage for School Name
overall_percent = overall_pass_count.groupby(["school_name"]).count()["student_name"] / total_stu_count * 100
overall_percent

# Create a dataframe to hold the above results
overview_summary = pd.DataFrame({"School Type": school_type,
                               "Total Students": total_stu_count,
                                 "Total School Budget": total_school_budget, 
                                 "Per Student Budget": per_stu_budget, 
                                 "Average Math Score": avg_math_score, 
                                 "Average Reading Score": avg_read_score,
                                 "% Passing Math": per_school_pass_math,
                                 "% Passing Reading": per_school_pass_read,
                                 "% Overall Passing": overall_percent})
overview_summary = overview_summary[["School Type", "Total Students", "Total School Budget", "Per Student Budget", "Average Math Score",
                                     "Average Reading Score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]                           
overview_summary["Total School Budget"] = overview_summary["Total School Budget"].map("${:,.2f}".format)
overview_summary["Per Student Budget"] = overview_summary["Per Student Budget"].map("${:,.2f}".format)
overview_summary
```

## Top Performing Schools (By % Overall Passing)

```python
# Sort and display the top five performing schools by % overall passing
top_sort = overview_summary.sort_values(["% Overall Passing"], ascending=False)
top_sort.head()
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#school_summary = pd.merge(overview_summary, school_data, how="left", on=["school_name", "school_name"])
#school_summary


# 4 reasonable bins to group school spending
spending_bins = [0, 579, 609, 639, 670]
group_labels = ["<$580", "$580-610", "$610-640", "$640-670"]
# Copying overview_summary to school_spending
school_spending = overview_summary
#  breaks down school performances based on average Spending Ranges (Per Student)
school_spending["Spending Range Per Student"] = pd.cut(per_stu_budget, spending_bins, labels=group_labels)
school_spending
school_spending_grouped = school_spending.groupby(["Spending Range Per Student"])
# Average Math Score based on spending range per student
math_avg = school_spending_grouped.mean()["Average Math Score"]
# Average Reading Score based on spending range per student
reading_avg = school_spending_grouped.mean()["Average Reading Score"]
# Pencentage of Average Reading Score based on spending range per student
percent_math_avg = school_spending_grouped.mean()["% Passing Math"]
# percentage of Average Reading Score based on spending range per student
percent_reading_avg = school_spending_grouped.mean()["% Passing Reading"]
# Percentage of Overall passing score based on spending range per student
percent_passing_overall = school_spending_grouped.mean()["% Overall Passing"]
spending_summary = pd.DataFrame({"Average Math Score": math_avg.round(2),
                                "Average Reading Score": reading_avg.round(2),
                                "% Passing Math": percent_math_avg.round(2),
                                "% passing Reading": percent_reading_avg.round(2),
                                "% Overall Passing": percent_passing_overall.round(2)})
spending_summary
```
